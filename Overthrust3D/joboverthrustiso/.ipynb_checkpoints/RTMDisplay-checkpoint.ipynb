```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage, misc
import matplotlib
import matplotlib.font_manager
import sys
nx=801
ny=801
nz=187
nt=2501
dx=25
dt=0.00033

font = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 15,
}
matplotlib.rc('font', **font)
def readfile(filename,nx,ny,nz):
    f=open(filename)
    data=np.fromfile(f,dtype='float32')
    data=data.reshape((nx,ny,nz))
    return data

def imagesc(data,scale,cmap='jet'):
    fig, ax = plt.subplots()
    nx,nz=np.shape(data)
    plt.imshow(data.T,vmin=scale*np.min(data),vmax=scale*np.max(data),extent=(0,nx*dx,0,nz*dx),aspect='auto', origin='lower',cmap=cmap)
    plt.colorbar()
    plt.gca().invert_yaxis() 
    plt.xlabel('distance(m)',fontsize=10)
    plt.ylabel('depth(m)',fontsize=10)
    ax.xaxis.tick_top()
#     plt.show()    
```

```python
plt.rcParams["figure.figsize"] = [10,4]
snap1=readfile('snap/shot1it20.dat',nx,ny,nz)
snap2=readfile('snap/shot1it20.dat-Reconstruction',nx,ny,nz)
# snap=readfile('snap/shot1it1000.dat-backward',nx,ny,nz)
imagesc(snap1[300:500,400,:50],1)
imagesc(snap2[300:500,400,:50],1)
plt.plot(snap1[401,400,:50],'b')
plt.plot(snap2[401,400,:50],'r')
```
